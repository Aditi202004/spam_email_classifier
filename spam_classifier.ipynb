```python
test_files = os.listdir(test_folder)
```

```python
train_emails = [folder + file for file in files]
test_emails = [test_folder + file for file in test_files]
```

```python
test_emails[0]
```

```python
words = []

for email in train_emails:
    f = open(email, encoding='utf-8')
    blob = f.read()
    words+=blob.split(" ")
```

```python
len(words)
```

```python
for i in range(len(words)):
    if not words[i].isalpha():
        words[i] = ""
```

```python
# tells the frequency of particular word
word_dict = Counter(words)
word_dict
```

```python
len(word_dict) # total unique words
```

```python
del word_dict[""]
```

```python
# we will only keep 3000 top words because other words haven't been used that frequently to affect
word_dict = word_dict.most_common(3000)
```

```python
len(word_dict)
```

```python
word_dict #list of tuples after using .most_common method
```

```python
# forming table
features = []
labels = []
for email in train_emails: # forming a row for each email
    f = open(email, encoding='utf-8')
    blob = f.read().split()
    row = []
    
    for i in word_dict:
        row.append(blob.count(i[0])) # i[0] because i is a tuple
    features.append(row)
    
    if 'spm' in email:
        labels.append(1)
    else:
        labels.append(0)
```
